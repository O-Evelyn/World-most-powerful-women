```python
# import the packages you will be using.
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
#your dataset is read into the programme.
p_women= pd.read_csv("C:\\Users\\49152\\Desktop\\The Worlds 100 Most Powerful Women.csv")
```

```python
#check to see if the import was successful and what your data looks like
p_women.head()
```

```python
#check if the data was completely imported
p_women.tail()
```

```python
#visualising the distribution of your data
fig, ax = plt.subplots(figsize=(12, 8))
scatter = ax.bar(p_women['AGE'],p_women['LOCATION'])
plt.xticks(rotation=45)
plt.show()
```

```python
#dataset has some missing values in AGE which did not appear as null values
# because they were inputed as (-), this code converts white space to NaN

p_women = p_women.replace(r'^\-*$', np.nan, regex=True)
```

```python
# checking for null values
p_women['AGE'].isnull().sum()
```

```python
#using the fillna function, the PAD method,where the null value are replaced with values in front of them.
p_women= p_women.fillna(method='ffill')
```

```python
# checking if all Null values have been replaced
p_women.isnull().sum()
```

```python
#saved the data back to my local computer
p_women.to_csv(r'C:\Users\49152\Desktop\Powerful_women.csv', index=False, header=True)
```

```python
#drop other rows to focus on USA which has the highest entry
women_usa= p_women[p_women.LOCATION == 'United States']
```

```python
women_usa.head(35)
```

```python
#creating a new column to group the ages of the women
# integers in the bin are boundaries e.g first group will be from 0-44,second group from 45-66 etc
# the bins have to be one position longer than the labels(bins had 5 instances, labels have 4)
p_women['AGE'] = p_women['AGE'].astype('int')
bins=[0,45,65,75,np.inf] 
labels = ['under 45','under 65','under 75','75+']
p_women['AGEGROUP'] = pd.cut(p_women['AGE'], bins=bins, labels=labels, right=False)
```
